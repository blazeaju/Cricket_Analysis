```python
# List all columns in the DataFrame to confirm if 'total' and 'id' are present
print(df.columns)

# If 'total' and 'id' are present, proceed with the drop operation
if 'total' in df.columns and 'id' in df.columns:
    X= df.drop(labels=['total','id'], axis=1)
    y= df['total'].values
else:
    print("Columns 'total' and/or 'id' not found in the DataFrame.")
    # Investigate further why these columns are missing
```

```python
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42,stratify=y)
```

```python
X_train= X_train.values
X_test=X_test.values
X_train=np.asarray(X_train).astype(np.float32)
X_test=np.asarray(X_test).astype(np.float32)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Handle missing values in X_train before fitting the model
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Assuming X_train and y_train are already defined

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)

# Fit the model with the imputed data
LR_model = LinearRegression()
LR_model.fit(X_train_imputed, y_train)
```

```python
#Creating a pickle file for the classifier
filename= 'lr-model.pkl'
pickle.dump(LR_model, open(filename, 'wb'))
```

```python
# Handle missing values in X_test before prediction
X_test_imputed = imputer.transform(X_test)  # Use the same imputer fitted on X_train

# Now predict using the imputed X_test
prediction = LR_model.predict(X_test_imputed)
mean_absolute_error(y_test, prediction)
```

```python
# Function to predict scores :

def score_prediction(Bat_team,Bowl_team,overs,total_score,total_wickets,run_rate,prev_runs_30,prev_wickets_30,prev_30_dot_balls,prev_30_boundaries,prev_30_run_rate):

    temp_array = list()

    # Add a placeholder for all teams (10 in total including 'other')
    batting_teams = ['AUS', 'BAN', 'ENG', 'IND', 'NZ', 'PAK', 'SA', 'SL', 'WI', 'Other']
    bowling_teams = ['AUS', 'BAN', 'ENG', 'IND', 'NZ', 'PAK', 'SA', 'SL', 'WI', 'Other']

    # One-hot encode batting team
    bat_team_encoded = [1 if team == Bat_team else 0 for team in batting_teams]
    temp_array.extend(bat_team_encoded)

    # One-hot encode bowling team
    bowl_team_encoded = [1 if team == Bowl_team else 0 for team in bowling_teams]
    temp_array.extend(bowl_team_encoded)

    temp_array = temp_array + [overs,total_score,total_wickets,run_rate,prev_runs_30,prev_wickets_30,prev_30_dot_balls,prev_30_boundaries,prev_30_run_rate]
    data = np.array([temp_array])
    my_prediction = int(LR_model.predict(data))
```
